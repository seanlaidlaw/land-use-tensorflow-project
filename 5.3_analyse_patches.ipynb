## Manually designed models

### Simple 2D CNN Model

```
model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(9,9,3)),
        Flatten(), # flatten into 1-D tensor
        Dense(6, activation='softmax') #output layer that categorizes images
])
model.compile(Adam(lr=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```
model.fit(x=all_patches_array, y=all_patches_ids, batch_size=128, epochs=10, validation_split=0.2)
```

### Comparison with Anti-overfit measures

```
model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(9,9,3)),
        Conv2D(32, (3, 3), activation='relu', padding="same"),
        keras.layers.MaxPooling2D(),

        Conv2D(32, (3, 3), activation='relu', padding="same"),
        Conv2D(32, (3, 3), activation='relu', padding="same"),
        keras.layers.MaxPooling2D(),

        keras.layers.BatchNormalization(),
        Flatten(), # flatten into 1-D tensor
        keras.layers.Dropout(0.5),
        Dense(6, activation='softmax') #output layer that categorizes images
])
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```
model.fit(x=all_patches_array, y=all_patches_ids, batch_size=128, epochs=10, validation_split=0.2)
```

## Transfer Learning

```
vgg16_model = keras.applications.vgg16.VGG16(include_top=True)
```

```
vgg16_model.summary()
```

```
model = Sequential()
for layer in vgg16_model.layers:
    model.add(layer)
```

```
model.summary()
```

```
NUM_TRAINABLE_LAYERS = 10
for layer in model.layers[:-NUM_TRAINABLE_LAYERS]:
    layer.trainable = False #excludes layer from future training so weights will not be updated when we train the model on our own data
```

```
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(6, activation='softmax'))
```

```
model.compile(Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```
